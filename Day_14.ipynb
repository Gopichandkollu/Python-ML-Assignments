```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"/content/Day_14_Pharma_data.csv")
df
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
df
```

```python
average_effectiveness_by_region = df.groupby(['Region', 'Product_ID'])['Effectiveness'].mean().unstack()
average_effectiveness_by_region.plot(kind='bar')
plt.title('Average Effectiveness for Each Drug Across Different Regions')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df)
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product')
plt.ylabel('Effectiveness')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_ID', y='Side_Effects', data=df)
plt.title('Distribution of Side Effects for Each Product')
plt.xlabel('Product')
plt.ylabel('Side Effects')
plt.show()
```

```python
sns.pairplot(df, vars=['Effectiveness', 'Side_Effects', 'Marketing_Spend'], hue='Product_Name')
plt.title('Relationships between Effectiveness, Side Effects, and Marketing Spend')
plt.show()
```

```python
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness for Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()
```

```python
sns.lmplot(x='Marketing_Spend', y='Effectiveness', data=df)
plt.title('Effect of Marketing Spend on Drug Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
